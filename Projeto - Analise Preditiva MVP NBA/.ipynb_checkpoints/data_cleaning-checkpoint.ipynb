```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mvps = pd.read_csv("mvps.csv")
```

```python
mvps
```

```python
mvps = mvps[["Player", "Year", "Pts Won", "Pts Max", "Share"]]
mvps.head()
```

```python
players = pd.read_csv("players.csv")
players
```

```python
del players["Unnamed: 0"]
del players["Rk"]

players.head()
```

```python
players["Player"].head(50)
```

```python
players["Player"] = players["Player"].str.replace("*", "", regex=False)
```

```python
players["Player"].head(50)
```

```python
players.head(20)
```

```python
players.groupby(["Player", "Year"]).get_group(("Greg Anderson", 1991)) #para verificar o agrupamento
```

```python
def single_row(df):
    if df.shape[0] == 1:
        return df
    else:
        row = df[df["Tm"] == "TOT"]
        row["Tm"] = df.iloc[-1,:]["Tm"]
        return row
    
players = players.groupby(["Player", "Year"]).apply(single_row)
```

```python
players.head(20)
```

```python
players.index = players.index.droplevel() #para remover o nivel de jogadores
players.index = players.index.droplevel() #para remover o nivel de ano
```

```python
players[players["Player"] == "Greg Anderson"]
```

```python
combined = players.merge(mvps, how="outer", on=["Player", "Year"])
```

```python
combined.head(5)
```

```python
# combined[combined["Pts Won"] > 0]
```

```python
combined[["Pts Won", "Pts Max", "Share"]] = combined[["Pts Won", "Pts Max", "Share"]].fillna(0)
```

```python
combined
```

```python
combined[combined["Pts Won"] > 0]
```

```python
teams = pd.read_csv("teams.csv")
teams.head(30)
```

```python
teams = teams[~teams["W"].str.contains("Division")]
```

```python
teams.head(30)
```

```python
teams["Team"] = teams["Team"].str.replace("*", "", regex=False)
teams.head(5)
```

```python
teams["Team"].unique()
```

```python
combined["Tm"].unique()
```

```python
nicknames = {}

with open("nicknames.csv") as f:
    lines = f.readlines()
    for line in lines[1:]:
        abbrev, name = line.replace("\n", "").split(",")
        nicknames[abbrev] = name
```

```python
combined["Team"] = combined["Tm"].map(nicknames)
```

```python
combined.head(5)
```

```python
stats = combined.merge(teams, how="outer", on=["Team", "Year"])
```

```python
stats[stats["Pts Won"] > 0]
```

```python
del stats["Unnamed: 0"]
```

```python
stats.dtypes
```

```python
stats = stats.apply(pd.to_numeric, errors="ignore") #vai por cada coluna do data frame e tenta converter para numerico
```

```python
stats.dtypes
```

```python
stats["GB"].unique()
```

```python
stats["GB"] = stats["GB"].str.replace("—", "0")
stats["GB"].unique()
```

```python
stats["GB"] = pd.to_numeric(stats["GB"])
stats.dtypes
```

```python
stats.to_csv("player_mvp_stats.csv")
```

```python
# criando um top 10 de maiores pontuadores, para jogadores com mais de 70 jogos
highest_scoring = stats[stats["G"] > 70].sort_values("PTS", ascending=False).head(10)
```

```python
highest_scoring.plot.bar("Player", "PTS")
```
